# 선형 회귀 직접 구현

```python
import numpy as np

# 임의의 데이터 생성
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # 100개의 샘플, 1개의 특징 (2D 배열)
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3X + 잡음
```

```python
# 경사 하강법 파라미터
learning_rate = 0.01
n_iterations = 1000
m = len(X)

# 랜덤 초기화
W = np.random.randn(1)
B = np.random.randn(1)

# 경사 하강법 구현
for i in range(n_iterations):
    # 예측값 계산
    y_pred = W * X + B

    # 비용 함수의 기울기 계산
    dW = (2/m) * np.dot(X.T, (y_pred - y))  # W에 대한 기울기
    dB = (2/m) * np.sum(y_pred - y)         # B에 대한 기울기

    # 파라미터 업데이트
    W = W - learning_rate * dW
    B = B - learning_rate * dB

print(f"학습된 가중치 W: {W[0]}")
print(f"학습된 절편 B: {B[0]}")
```

```python
import matplotlib.pyplot as plt

# 예측값 계산
y_pred = W * X + B

# 시각화
plt.scatter(X, y, color="blue", label="Actual Data")  # 실제 데이터
plt.plot(X, y_pred, color="red", label="Fitted Line")  # 학습된 선형 회귀선
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

# 선형 회귀 구현

```python
import numpy as np

# 데이터 생성
np.random.seed(0)  # 재현성을 위해 난수 시드 설정
X = 2 * np.random.rand(100, 1)  # 100개의 샘플, 1개의 특징
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3X + 잡음
```

```python
from sklearn.linear_model import LinearRegression

# 모델 초기화
model = LinearRegression()

# 모델 학습
model.fit(X, y)
```

```python
# 예측
y_pred = model.predict(X)

# 학습된 모델의 가중치와 절편 출력
print(f"학습된 가중치 (W): {model.coef_[0][0]}")
print(f"학습된 절편 (B): {model.intercept_[0]}")
```

```python
import matplotlib.pyplot as plt

# 시각화
plt.scatter(X, y, color="blue", label="Actual Data")  # 실제 데이터
plt.plot(X, y_pred, color="red", label="Fitted Line")  # 예측된 선형 회귀선
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
